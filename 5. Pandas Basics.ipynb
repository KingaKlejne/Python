```python
import os
import glob
import pandas as pd
```

```python
#tworzenie sciezek dla plikow data.xls i data_more_observations.xls
file_path =  "/Users/kingaklejne/Desktop/0. KURS/jdszr3-materialy-python/8_data_science_tools/dummy_data/data_continued/"
file_name_data = "data.xls"
file_name_more = "data_more_observations.xls"

file_path_data = os.path.join(file_path, file_name_data)
file_path_more = os.path.join(file_path, file_name_more)
```

```python
#ramka pliku data.xls
df_data = pd.read_excel(file_path_data)
df_data
```

```python
##ramka pliku data_more_observations.xls
df_more = pd.read_excel(file_path_more)
df_more
```

```python
#połączenie ramek data.xls i data_more_observations.xls
df = pd.concat([df_data, df_more])
df.reset_index(drop=True)
```

```python
#sprawdzenie czy są duplikaty w kolumnie 'id' - są
czy_duplikat = df[['id']].duplicated()
czy_duplikat.any()
```

```python
#liczenie ile jest duplikatow - 24
t = 0

for i in czy_duplikat:
    if i == True:
        t += 1
        
print(t)
```

```python
#proba usuniecia duplikatow z kolumny 'id'
temp_1 = df.copy()
temp_1[['id']].drop_duplicates()
```

```python
#usuniecie duplikatow z kolumny 'id' na nowym dataframe, zeby nie modyfikowac orginalow
temp_1 = temp_1.drop_duplicates(subset=["id"], keep="last").reset_index(drop=True)
```

```python
#przedstawienie nowej ramki danych
temp_1
```

```python
#tworzenie sciezki dla pliku data_more_features.xls
file_name_features = "data_more_features.xls"
file_path_features = os.path.join(file_path, file_name_features)
```

```python
#tworzenie ramki dla pliku data_more_features.xls
df_features = pd.read_excel(file_path_features)
df_features
```

```python
#mergowanie wczesniejszej ramki z data_more_features.xls
df_new = pd.merge(temp_1, df_features, on="id", how="inner")
df_new
```

```python
#sprawdzenie czy nie ma wartosci NaN
missing_data = df_new.isnull().sum()
missing_data
```
