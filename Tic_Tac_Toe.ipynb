```python
board = [' ',' ',' ',' ',' ',' ',' ',' ',' ']

def display_board(board):
    print(board[0] + '|' + board[1] + '|' + board[2])
    print('-+-+-')
    print(board[3] + '|' + board[4] + '|' + board[5])
    print('-+-+-')
    print(board[6] + '|' + board[7] + '|' + board[8])
    return 'Tic Tac Toe'

turn = 'O'
count = 0
stage = 'game continues'

game = input('Do you want to play in Tic Tac Toe?\n')
print(display_board(board))

while game.lower() == 'yes':
    
    move = int(input('\nYour turn ' + turn + '. Move to which place? You can choose 0 - 8. '))
    if 0 <= move <= 8:
        if board[move] == ' ':
            board[move] = turn
            print(display_board(board))
            count += 1
        else:
            print('\nThat place is already filled. Move to which place?')
            continue
    else:
        print('Invalid number')


    if count >= 5:
        if board[0] == board[1] == board[2] != ' ':
            stage = 'END OF THE GAME!'
            print('\n' + stage + '\n' + turn + ' won.')
            pass
        elif board[3] == board[4] == board[5] != ' ':
            stage = 'END OF THE GAME!'
            print('\n' + stage + '\n' + turn + ' won.')
            pass
        elif board[6] == board[7] == board[8] != ' ':
            print(board)
            stage = 'END OF THE GAME!'
            print('\n' + stage + '\n' + turn + ' won.')
            pass
        elif board[0] == board[3] == board[6] != ' ':
            stage = 'END OF THE GAME!'
            print('\n' + stage + '\n' + turn + ' won.')
            pass
        elif board[1] == board[4] == board[7] != ' ':
            stage = 'END OF THE GAME!'
            print('\n' + stage + '\n' + turn + ' won.')
            pass
        elif board[2] == board[5] == board[8] != ' ':
            stage = 'END OF THE GAME!'
            print('\n' + stage + '\n' + turn + ' won.')
            pass
        elif board[0] == board[4] == board[8] != ' ':
            stage = 'END OF THE GAME!'
            print('\n' + stage + '\n' + turn + ' won.')
            pass
        elif board[2] == board[4] == board[6] != ' ':
            stage = 'END OF THE GAME!'
            print('\n' + stage + '\n' + turn + ' won.')
            pass
        elif count == 9:
            stage = 'END OF THE GAME!'
            print('\n' + stage + '\n' + 'DRAW!')

    if turn =='O':
        turn = 'X'
    else:
        turn = 'O'

    if stage == 'END OF THE GAME!':
        dalej = input('Do you want to play again? ')
        if dalej.lower() == 'yes':
            board = [' ',' ',' ',' ',' ',' ',' ',' ',' ']
            count = 0
            stage = 'game continues'
            continue
        else:
            print('\nBye Bye! :)')
            break


else:
    print('\nBye Bye! :)')
    
```
